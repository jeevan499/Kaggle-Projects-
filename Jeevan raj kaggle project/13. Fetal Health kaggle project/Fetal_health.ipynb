# Model Building 

## Decision Tree


###### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
DT_df = df.copy()
DT_df
```

```python
DT_df['prolongued_decelerations'] = LE.fit_transform(DT_df['prolongued_decelerations'])
DT_df['percentage_of_time_with_abnormal_long_term_variability'] = LE.fit_transform(DT_df['percentage_of_time_with_abnormal_long_term_variability'])
DT_df['histogram_tendency'] = LE.fit_transform(DT_df['histogram_tendency'])
DT_df['fetal_health'] = LE.fit_transform(DT_df['fetal_health'])
```

```python
DT_df
```

```python
#Importing algorithm
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
```

```python
X = DT_df.loc[:, DT_df.columns != 'fetal_health']
```

```python
y = DT_df.loc[:, DT_df.columns == 'fetal_health']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state= 14)
```

```python
dt_clf = DecisionTreeClassifier() 
```

```python
#Fit this classifier model on train data set
dt_clf.fit(X_train, y_train)
```

```python
y_pred_train_dt = dt_clf.predict(X_train)
y_pred_train_dt
```

```python
y_pred_test_dt = dt_clf.predict(X_test)
y_pred_test_dt
```

```python
# see the tree
from sklearn import tree
tree.plot_tree(dt_clf.fit(X_train, y_train))
```

```python
# Model has learnt everything that is nothing but overfitting
# we need to optimize

from sklearn.metrics import accuracy_score
```

```python
print(round(accuracy_score(y_train, y_pred_train_dt),2))
print(round(accuracy_score(y_test, y_pred_test_dt),2))
```

```python
from sklearn import tree
path = dt_clf.cost_complexity_pruning_path(X_train, y_train)
```

```python
alphas = path['ccp_alphas']
alphas
```

```python
accuracy_train, accuracy_test = [],[]
```

```python
for i in alphas:
    dt_clf = DecisionTreeClassifier(ccp_alpha=i)    
    
    dt_clf.fit(X_train, y_train)
    y_pred_train1 = dt_clf.predict(X_train)
    y_pred_test1 = dt_clf.predict(X_test)
    
    accuracy_train.append(accuracy_score(y_train, y_pred_train1))
    accuracy_test.append(accuracy_score(y_test,y_pred_test1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.30,random_state =14)
```

```python
print("length of x_train:",len(x_train)) 
print("length of x_test:",len(x_test)) 
print("length of y_train:",len(y_train))
print("length of y_test:",len(y_test)) 
```

#### Using 'for-loop' to find optimum n_neighbor value

```python
neighbors = np.arange(1, 30)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
# Using "for loop" to find n_neghbor value
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
     
    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test)
```

```python
 
#We have the accuracy, lets plot
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
 
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

**We can see from the graph that @n_neighbors = 26 we are getting 
the highest accuracy in test dataset and and the minimum difference 
between train accuray and test accuracy.**

**Therefore, the optimum value of K is 26.**

###### Building Model with obtained @ n_neighbors = 26
