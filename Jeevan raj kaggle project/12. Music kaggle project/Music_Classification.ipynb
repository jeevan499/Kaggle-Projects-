```python
#Outlier treatment using 'for-loop' with lambda
for i in col_names:
    q1, q2, q3 = df1[i].quantile([0.25,0.5,0.75])
    IQR = q3 - q1
    LL=q1-1.5*IQR
    UL=q3+1.5*IQR
    df1[i]=df1[i].apply(lambda x: UL if x>(UL) else (LL if x<(LL) else x))
```

```python
#For Outliers treatment we can use user feined function
#This cell and the next is the procedure for outliers treatment
def remove_outliers(data):
    arr=[]
    #print(max(list(data)))
    q1=np.percentile(data,25)
    q3=np.percentile(data,75)
    iqr=q3-q1
    mi=q1-(1.5*iqr)
    ma=q3+(1.5*iqr)
    #print(mi,ma)
    for i in list(data):
        if i<mi:
            i=mi
            arr.append(i)
        elif i>ma:
            i=ma
            arr.append(i)
        else:
            arr.append(i)
    #print(max(arr))
    return arr
```

#### Using 'for-loop' to find optimum n_neighbor value

```python
neighbors = np.arange(1, 30)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
# Using "for loop" to find n_neghbor value
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
     
    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
 
#We have the accuracy, lets plot
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
 
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
''' We can see from the graph that @n_neighbors = 8 we are getting 
the highest accuracy in test dataset.
Therefore, the optimum value of K is 8.'''
```
